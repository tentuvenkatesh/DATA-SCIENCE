# code for iris dataset:

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score


#1.Load the Iris dataset
iris = load_iris()
X, y = iris.data[:, :2], iris.target

#2.Visualizing the data
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Dataset: Sepal Length vs Sepal Width')
plt.show()

#3.Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#4.Initializing and training the SVM model
# Initialize SVM classifier
svm_clf = SVC(kernel='linear', C=1.0, random_state=42)
# Train the SVM model
svm_clf.fit(X_train, y_train)

#5.Test the model
y_pred = svm_clf.predict(X_test)

# Print classification report
print("Classification Report for Iris Dataset:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy on test set: {accuracy:.2f}')
```

# code for digits dataset:

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score


#1.Load the digits dataset
digits = load_digits()
X, y = digits.data, digits.target

#2.Visualize some digits
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(10, 4))

for ax, image, label in zip(axes.flatten(), digits.images, digits.target):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title(f'Target: {label}')

plt.tight_layout()
plt.show()

#3.Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#4.Initializing and training the SVM model
# Initialize SVM classifier
svm_clf = SVC(kernel='rbf', C=10, gamma=0.001, random_state=42)
# Train the SVM model
svm_clf.fit(X_train, y_train)

#5.Test the model
y_pred = svm_clf.predict(X_test)

#Classification Report
print("Classification Report for Digits Dataset:")
print(classification_report(y_test, y_pred))

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy on test set: {accuracy:.2f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the iris dataset from sklearn
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Standardize the data
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# Perform PCA using numpy's Eigen decomposition
cov_matrix = np.cov(X_std.T)
eigen_values, eigen_vectors = np.linalg.eig(cov_matrix)

# Sort eigenvalues and eigenvectors in descending order
sorted_indices = np.argsort(eigen_values)[::-1]
eigen_values_sorted = eigen_values[sorted_indices]
eigen_vectors_sorted = eigen_vectors[:, sorted_indices]

# Project the data onto the new feature space
n_components = 3  # Number of components to keep
projection_matrix = eigen_vectors_sorted[:, :n_components]
X_pca = X_std.dot(projection_matrix)

# Compare with scikit-learn's PCA results
pca_sklearn = PCA(n_components=3)
X_pca_sklearn = pca_sklearn.fit_transform(X_std)

# Plot the PCA results side by side
fig, axes = plt.subplots(1, 2, figsize=(6, 6), subplot_kw={'projection': '3d'})

# Plot using numpy's PCA
axes[0].set_title('PCA using numpy Eigen decomposition')
for target, color in zip(range(len(iris.target_names)), ['r', 'g', 'b']):
    axes[0].scatter(X_pca[y == target, 0], X_pca[y == target, 1], X_pca[y == target, 2], color=color, label=iris.target_names[target])
axes[0].set_xlabel('PC1')
axes[0].set_ylabel('PC2')
axes[0].set_zlabel('PC3')
axes[0].legend(loc='best')

# Plot using scikit-learn's PCA
axes[1].set_title('PCA using scikit-learn')
for target, color in zip(range(len(iris.target_names)), ['r', 'g', 'b']):
    axes[1].scatter(X_pca_sklearn[y == target, 0], X_pca_sklearn[y == target, 1], X_pca_sklearn[y == target, 2], color=color, label=iris.target_names[target])
axes[1].set_xlabel('PC1')
axes[1].set_ylabel('PC2')
axes[1].set_zlabel('PC3')
axes[1].legend(loc='best')

plt.tight_layout()
plt.show()

# Compare the results
print("First 5 samples of PCA using numpy Eigen decomposition:")
print(X_pca[:5])
print("\nFirst 5 samples of PCA using scikit-learn:")
print(X_pca_sklearn[:5])
```
